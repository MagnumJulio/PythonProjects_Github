```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("brutos/microdados_enem_2021/DADOS/MICRODADOS_ENEM_2021.csv", sep=";", encoding='ISO-8859-1', nrows=9999999)
#df.columns.values

lista = ['NU_INSCRICAO', 'TP_FAIXA_ETARIA', 'TP_SEXO',
       'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_NACIONALIDADE',
       'IN_TREINEIRO', 'CO_MUNICIPIO_ESC']

df_filt = df.filter(items=lista)

df_filt.columns.values

#df_filt[0:900]
dicto = df_filt['TP_FAIXA_ETARIA']


faixa_etaria = list()

#plt.bar(dicto[], )
for age in dicto.value_counts().index:
       #print(age)
       faixa_etaria.append(str(age))
       

#dicto.value_counts().values


nums = dicto.value_counts().values

#dicto.value_counts()
#print(valores)
#dicto.value_counts()

#df.head()
```
